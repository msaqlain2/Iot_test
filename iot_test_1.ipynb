```python
#Question # 4

numbers = []

user_input = int(input("Enter No. of Elemenets "))


for i in range(0, user_input):
  appendelemets = int(input())
  numbers.append(appendelemets)

copy_of_integers = numbers[:]  

largest_integer = max(copy_of_integers)  

print("Largest Number is", largest_integer)

copy_of_integers.remove(largest_integer)

second_largest_integer = max(copy_of_integers)  

print("Seond Largest Number is ",second_largest_integer)

```

```python
#Question # 9

number = int(input("Enter a number: "))

factorial = 1

if number < 0:
   print("Sorry, factorial does not exist for negative numbers")
elif number == 0:
   print("The factorial of 0 is 1")
else:
   for i in range(1,number + 1):
       factorial = factorial*i
   print("The factorial of",number,"is",factorial)
```
